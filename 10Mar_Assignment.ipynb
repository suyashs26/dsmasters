```python
#2
import math

def estimate_population_mean(sample_mean, sample_std, n):
    standard_error = sample_std / math.sqrt(n)
    margin_of_error = 1.96 * standard_error # assuming 95% confidence level
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    estimated_population_mean = (lower_bound + upper_bound) / 2
    return estimated_population_mean
```

```python
#9
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10  # degrees of freedom

x = np.linspace(-4, 4, 1000)  # range of x values
y = t.pdf(x, df)  # calculate y values for t-distribution

plt.plot(x, y, label='t-distribution')
plt.legend()
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
```

```python
#10
import numpy as np
from scipy.stats import ttest_ind

# Generate two random samples of equal size
sample1 = np.random.normal(loc=5, scale=2, size=50)
sample2 = np.random.normal(loc=6, scale=2, size=50)

# Calculate the t-test statistic and p-value
t_stat, p_val = ttest_ind(sample1, sample2, equal_var=True)

# Print the results
print(f"T-test statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

```
