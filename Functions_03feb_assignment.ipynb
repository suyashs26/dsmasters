```python
#1
#def keyword is used to create function in python
def odd_numbers():
    l = []
    for i in range(1,25):
        if(i%2!=0):
            l.append(i)
    return l        
            
print(odd_numbers())            
```

```python
#2
#We use args and kwargs to pass variable number of arguments to a function. It is used when we are unsure of arguments required in function
# *args : non keyword argument ; Eg:
def adder(*num):
    sum = 0   
    for n in num:
        sum = sum + n
    return sum
print(adder(3,5))
print(adder(4,5,8))

# **kwargs:  keyword argument ; Eg:
def test1(**kwargs):
    return kwargs
test1(a=1,b=2,c=[1,2,4],d={'sk','ws'})
print(test1(a=1,b=2,c=[1,2,4],d={'sk','ws'})['d'])
    
```

```python
#4
# generator is a function that returns an iterator which produces a sequence of values when iterated over.
# yield keyword is used to create a generator function. A type of function that is memory efficient and can be used like an iterator object. 
def simple():  
    for i in range(10):  
        if(i%2==0):  
             yield i  
for i in simple():  
    print(i)  
```
